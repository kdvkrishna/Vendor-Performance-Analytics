```python
import pandas as pd
import os
from sqlalchemy import create_engine

# Connection
engine = create_engine('postgresql://postgres:gtsr@localhost:5432/vendor_performance')

# Ingestion 
def ingest_all_csvs(folder_path):
    # 1. Getting a list of all files in the 'data' folder
    files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    
    for file_name in files:
        # Creating a clean table name 
        table_name = file_name.replace('.csv', '')
        file_path = os.path.join(folder_path, file_name)
        
        print(f"Starting ingestion for: {table_name}...")
        
        # 2. Setting up the "Chunk" reader for 100,000 rows 
        reader = pd.read_csv(file_path, chunksize=100000)
        
        # 3. Looping through the chunks and pushing them to Postgres
        for i, chunk in enumerate(reader):
            # If it's the first chunk, 'replace' the old table
            # If it's chunk 2, 3, 4..., 'append' to the end of the table
            if_exists_strategy = 'replace' if i == 0 else 'append'
            
            chunk.to_sql(table_name, con=engine, if_exists=if_exists_strategy, index=False)
            
            # Print progress
            if i % 5 == 0:
                print(f"   Processed {i * 100000} rows for {table_name}...")

        print(f" {table_name} successfully loaded!\n")


    # RUN THE SCRIPT
if __name__ == "__main__":
    ingest_all_csvs(r'E:\AAKDV\Vendor_Performance_Analysis SQL+PYTHON+BI\data')
```
